# Ministerio de Ciencia Tecnología e Innovación

```python
def data_minciencias():
    base_url = "https://minciencias.gov.co"
    url_objetivo = "https://minciencias.gov.co/convocatorias/todas"  
    print(f"{'='*20} MINCIENCIAS: {'='*20}")
    try:
        resp = requests.get(url_objetivo, timeout=10)
        soup = BeautifulSoup(resp.text, "html.parser")
        datos_minciencias = []
        filas = soup.find_all("tr")
        for fila in filas:
            cols = fila.find_all("td")
            if len(cols) >= 2:
                link_tag = cols[1].find("a")
                if link_tag and link_tag.has_attr('href'):
                    url_final = urljoin(base_url, link_tag['href'])
                    titulo = link_tag.get_text(strip=True)
                    print(f"Institucion: MinCiencias")
                    print(f"Titulo: {titulo}")
                    print(f"Url: {url_final}")
                    print("-" * 50)                  
                    datos_minciencias.append({"institucion": "MinCiencias", "titulo_convocatoria": titulo, "url_externa": url_final})       
        return datos_minciencias
    except Exception as e:
        print(f"Error al corregir MinCiencias: {e}")
        return []
minciencias = data_minciencias()
```

# Fundación Natura Colombia

```python
def data_natura():
    url_objetivo = "https://natura.org.co/convocatorias/"
    base_url = "https://natura.org.co"  
    print(f"{'='*20} FUNDACIÓN NATURA: {'='*20}") 
    try:
        headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36'}
        resp = requests.get(url_objetivo, headers=headers, timeout=10)
        soup = BeautifulSoup(resp.text, "html.parser")
        datos_natura = []
        enlaces_vistos = set()
        for a in soup.find_all("a", href=True):
            href = a['href']
            if "convocatoria" in href.lower() and href not in enlaces_vistos:
                url_final = urljoin(base_url, href)
                slug = urlparse(url_final).path.strip('/')
                titulo_limpio = slug.replace('-', ' ').lower()
                titulo_limpio = titulo_limpio.replace('convocatoria', '').strip().capitalize()       
                print(f"Institucion: Fundación Natura")
                print(f"Titulo (Decodificado): {titulo_limpio}")
                print(f"Url: {url_final}")
                print("-" * 50)
                datos_natura.append({"institucion": "Fundación Natura", "titulo_convocatoria": titulo_limpio, "url_externa": url_final})
                enlaces_vistos.add(href)
        return datos_natura
    except Exception as e:
        print(f"Error al corregir Fundación Natura: {e}")
        return []
fundacion_natura = data_natura()
```

# Fundación para la Conservación y el Desarrollo

```python
def data_fcds():
    url_objetivo = "https://fcds.org.co/oportunidades-laborales/"
    base_url = "https://fcds.org.co"
    print(f"{'='*20} FCDS: {'='*20}")
    try:
        headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36'}
        resp = requests.get(url_objetivo, headers=headers, timeout=10)
        soup = BeautifulSoup(resp.text, "html.parser")     
        datos_fcds = []
        enlaces_vistos = set()
        enlaces = soup.find_all("a", href=True)
        for a in enlaces:
            href = a['href']
            texto_boton = a.get_text(strip=True)
            if (".pdf" in href.lower() and "uploads" in href.lower()) or ("más información" in texto_boton.lower()):        
                if href not in enlaces_vistos:
                    url_final = urljoin(base_url, href)
                    nombre_archivo = urlparse(url_final).path.split('/')[-1]
                    titulo_decodificado = nombre_archivo.replace('.pdf', '').replace('-', ' ').replace('_', ' ')
                    titulo_decodificado = titulo_decodificado.replace('tdr', 'TDR:').strip().capitalize()          
                    print(f"Institucion: FCDS")
                    print(f"Titulo: {titulo_decodificado}")
                    print(f"Url: {url_final}")
                    print("-" * 50)
                    datos_fcds.append({"institucion": "FCDS","titulo_convocatoria": titulo_decodificado,"url_externa": url_final})
                    enlaces_vistos.add(href)
        return datos_fcds
    except Exception as e:
        print(f"Error al corregir FCDS: {e}")
        return []
fcds = data_fcds()
```

# FontAgro

```python
def data_fontagro():
    url_objetivo = "https://fontagro.org/es/iniciativas/convocatorias"
    base_url = "https://fontagro.org" 
    print(f"{'='*20} FONTAGRO: {'='*20}")
    try:
        headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36'}
        resp = requests.get(url_objetivo, headers=headers, timeout=10)
        soup = BeautifulSoup(resp.text, "html.parser")
        datos_fontagro = []
        tarjetas = soup.find_all("h4", class_="initiative-title") 
        for titulo_tag in tarjetas:
            titulo = titulo_tag.get_text(strip=True)   
            link_tag = titulo_tag.find("a") if titulo_tag.find("a") else titulo_tag.find_next("a")        
            if link_tag and link_tag.has_attr('href'):
                href = link_tag['href']
                url_final = urljoin(base_url, href)
                print(f"Institucion: FontAgro")
                print(f"Titulo: {titulo}")
                print(f"Url: {url_final}")
                print("-" * 50)
                datos_fontagro.append({"institucion": "FontAgro","titulo_convocatoria": titulo,"url_externa": url_final}) 
        if not datos_fontagro:
            print("Aviso: No se detectaron convocatorias con los selectores actuales. Verificando estructura...")
        return datos_fontagro
    except Exception as e:
        print(f"Error al corregir FontAgro: {e}")
        return []
fontagro = data_fontagro()
```

```python
def data_colombia_sostenible():
    url_objetivo = "https://www.colombiasostenible.gov.co/programa/convocatorias/"
    base_url = "https://www.colombiasostenible.gov.co" 
    print(f"{'='*20} COLOMBIA SOSTENIBLE: {'='*20}") 
    try:
        driver.get(url_objetivo)
        WebDriverWait(driver, 15).until(
            EC.presence_of_element_located((By.CLASS_NAME, "entry-title")))  
        soup = BeautifulSoup(driver.page_source, "html.parser")
        datos_colombia = []
        articulos = soup.find_all("h3", class_="entry-title")  
        if not articulos:
            articulos = soup.select(".entry-content a")
        for item in articulos:
            link_tag = item if item.name == "a" else item.find("a")
            if link_tag and link_tag.has_attr('href'):
                titulo = link_tag.get_text(strip=True)
                href = link_tag['href']    
                if len(titulo) > 10:
                    url_final = urljoin(base_url, href)  
                    print(f"Institucion: Colombia Sostenible")
                    print(f"Titulo: {titulo}")
                    print(f"Url: {url_final}")
                    print("-" * 50)
                    datos_colombia.append({"institucion": "Colombia Sostenible","titulo_convocatoria": titulo,"url_externa": url_final}) 
        return datos_colombia
    except Exception as e:
        print(f"Error al obtener datos de Colombia Sostenible: {e}")
        return []
colombia_sostenible = data_colombia_sostenible()
```

# Fondo Acción

```python
def data_fondo_accion():
    url_objetivo = "https://fondoaccion.org/convocatorias/"
    print(f"{'='*20} FONDO ACCIÓN: {'='*20}")    
    try:
        driver.get(url_objetivo)   
        WebDriverWait(driver, 15).until(
            EC.presence_of_element_located((By.CLASS_NAME, "fg-item-title")))
        soup = BeautifulSoup(driver.page_source, "html.parser")
        datos_fondo = []
        enlaces_vistos = set()
        titulos_bloques = soup.find_all("h2", class_="fg-item-title")
        for h2 in titulos_bloques:
            titulo = h2.get_text(strip=True)
            link_tag = h2.find_parent("a")
            if link_tag and link_tag.has_attr('href'):
                url_real = link_tag['href']
                if url_real not in enlaces_vistos and not url_real.lower().endswith(('.jpg', '.jpeg', '.png')):            
                    print(f"Institucion: Fondo Acción")
                    print(f"Titulo: {titulo}")
                    print(f"Url: {url_real}")
                    print("-" * 50)     
                    datos_fondo.append({"institucion": "Fondo Acción","titulo_convocatoria": titulo,"url_externa": url_real})
                    enlaces_vistos.add(url_real)
        return datos_fondo
    except Exception as e:
        print(f"Error al corregir Fondo Acción: {e}")
        return []
fondo_accion = data_fondo_accion()
```

# Banco de Desarrollo de América Latina (CAF)

```python
def data_caf():
    urls_objetivo = [
        "https://www.caf.com/es/trabaja-con-nosotros/convocatorias/",
        "https://www.caf.com/es/trabaja-con-nosotros/licitaciones/"
    ]
    base_url = "https://www.caf.com"
    print(f"{'='*20} CAF: (Convocatorias + Licitaciones) {'='*20}")
    datos_caf = []
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36'}
    for url in urls_objetivo:
        try:
            resp = requests.get(url, headers=headers, timeout=15)
            soup = BeautifulSoup(resp.text, "html.parser")
            encabezados_titulo = soup.find_all("h3", class_=["card__title", "card--gris-50__title"])        
            for h3 in encabezados_titulo:
                link_tag = h3.find("a", href=True)            
                if link_tag:
                    titulo = link_tag.get_text(strip=True)
                    href = link_tag['href']
                    url_final = urljoin(base_url, href)
                    if len(titulo) > 10 and "mailto:" not in url_final.lower():
                        print(f"Institucion: CAF")
                        print(f"Titulo: {titulo}")
                        print(f"Url: {url_final}")
                        print("-" * 50)            
                        datos_caf.append({"Institucion": "CAF","Titulo": titulo,"Url": url_final})              
        except Exception as e:
            print(f"Error procesando {url}: {e}")
    return datos_caf
caf = data_caf()
```

# GIZ (Deutsche Gesellschaft für Internationale Zusammenarbeit)

```python
def data_giz():
    url_objetivo = "https://www.giz.de/en/workingwithgiz/tenders.html"
    base_url = "https://www.giz.de"
    print(f"{'='*20} GIZ: {'='*20}")
    try:
        headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36'}
        resp = requests.get(url_objetivo, headers=headers, timeout=15)
        soup = BeautifulSoup(resp.text, "html.parser")
        datos_giz = []
        enlaces = soup.find_all("a", href=True)
        for link in enlaces:
            href = link['href']
            if any(term in href.lower() for term in ["tender", "procurement", "contract"]):
                titulo = link.get_text(strip=True)
                url_final = urljoin(base_url, href)
                ruido = ["read more", "click here", "details", "contact", "download"]
                if len(titulo) > 20 and not any(r in titulo.lower() for r in ruido):
                    if "mailto:" not in url_final.lower():
                        print(f"Institucion: GIZ")
                        print(f"Titulo: {titulo}")
                        print(f"Url: {url_final}")
                        print("-" * 50)
                        datos_giz.append({"Institucion": "GIZ","Titulo": titulo,"Url": url_final})
        return datos_giz
    except Exception as e:
        print(f"Error al procesar GIZ: {e}")
        return []
giz = data_giz()
```

# Extract information from URL

```python
import os
import time
import random
import shutil
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager

def crear_driver():
    """Configura e inicia una instancia limpia de Chrome."""
    chrome_options = Options()
    chrome_options.add_argument("--start-maximized")
    chrome_options.add_argument("--disable-dev-shm-usage")
    chrome_options.add_argument("--no-sandbox")
    # Importante: No usar headless para que el visor de PDF de Chrome funcione bien
    service = Service(ChromeDriverManager().install())
    return webdriver.Chrome(service=service, options=chrome_options)

def aplanar_a_ascii_puro(input_file, output_file):
    """Elimina cualquier carácter no ASCII."""
    print(f"🔨 Aplanando archivo a ASCII puro: {input_file}")
    try:
        with open(input_file, 'rb') as f_in:
            data = f_in.read()
        data_plana = bytes([b for b in data if b < 128])
        with open(output_file, 'wb') as f_out:
            f_out.write(data_plana)
        print(f"✅ Archivo 100% ASCII generado: {output_file}")
    except Exception as e:
        print(f"❌ Error en aplanado: {e}")

def ejecutar_proceso_gena_completo(df):
    archivo_sql = "universo_licitaciones.sql"
    archivo_ascii = "universo_licitaciones_ASCII.sql"
    tmp_dir = "temp_html_files"
    
    if not os.path.exists(tmp_dir):
        os.makedirs(tmp_dir)

    with open(archivo_sql, "w", encoding='utf-8') as f_sql:
        f_sql.write("-- SEI-GENA: Script de carga masiva\n")
        f_sql.write('CREATE SCHEMA IF NOT EXISTS "GENA_Schema";\n')
        f_sql.write('CREATE TABLE IF NOT EXISTS "GENA_Schema".licitaciones (institucion TEXT, titulo TEXT, url TEXT, informacion TEXT);\n')
        f_sql.write('TRUNCATE TABLE "GENA_Schema".licitaciones;\n\n')

    print(f"🚀 Iniciando extracción masiva: {len(df)} registros.")
    driver = crear_driver()
    
    try:
        for idx, row in df.iterrows():
            # Reiniciar driver cada 100 registros para estabilidad
            if idx > 0 and idx % 100 == 0:
                print(f"🔄 Reiniciando navegador (Registro {idx})...")
                driver.quit()
                driver = crear_driver()

            intentos = 0
            exito = False
            url_actual = str(row['Url'])
            
            while intentos < 2 and not exito:
                try:
                    # LÓGICA DE EXTRACCIÓN UNIVERSAL (HTML y PDF vía Navegador)
                    driver.get(url_actual)
                    time.sleep(6) # Tiempo para que cargue el PDF o el JS del sitio
                    
                    if url_actual.lower().endswith(".pdf") or "document.cfm" in url_actual.lower():
                        # Si es PDF, extraemos el texto del cuerpo del visor de Chrome
                        texto_limpio = driver.find_element(By.TAG_NAME, "body").text
                        if not texto_limpio or len(texto_limpio) < 20:
                            texto_limpio = f"Documento PDF detectado. Consultar URL: {url_actual}"
                    else:
                        # Si es HTML, usamos BeautifulSoup como siempre
                        html_raw = driver.page_source
                        soup = BeautifulSoup(html_raw, "html.parser")
                        for tag in soup(["script", "style", "nav", "footer", "header", "aside"]):
                            tag.decompose()
                        texto_limpio = soup.get_text(separator=' ', strip=True)
                        
                        # Guardar respaldo
                        with open(os.path.join(tmp_dir, f"registro_{idx}.html"), "w", encoding='utf-8') as f_h:
                            f_h.write(html_raw)

                    # Preparación SQL (Escape de comillas)
                    info_sql = texto_limpio.replace("'", "''")
                    inst_sql = str(row['Institucion']).replace("'", "''")
                    tit_sql = str(row['Titulo']).replace("'", "''")
                    url_sql = url_actual.replace("'", "''")

                    # Escritura con columnas explícitas para evitar warnings
                    with open(archivo_sql, "a", encoding='utf-8') as f_sql:
                        query = f"INSERT INTO \"GENA_Schema\".licitaciones (institucion, titulo, url, informacion) VALUES ('{inst_sql}', '{tit_sql}', '{url_sql}', '{info_sql}');\n"
                        f_sql.write(query)
                        f_sql.flush()

                    print(f"✅ [{idx+1}/{len(df)}] Procesado: {row['Institucion'][:15]}...")
                    exito = True

                except Exception as e:
                    print(f"⚠️ Error registro {idx} (Intento {intentos+1}): {e}")
                    intentos += 1
                    driver.quit()
                    driver = crear_driver()

    finally:
        driver.quit()
        if os.path.exists(tmp_dir):
            shutil.rmtree(tmp_dir)
        
        # Generar versión ASCII pura
        aplanar_a_ascii_puro(archivo_sql, archivo_ascii)
        print(f"\n✨ PROCESO TERMINADO ✨")

# Ejecución
ejecutar_proceso_gena_completo(df_unificado)
```

# Import Large Languaje Models (LLMs) and Transformers (NNT)

```python
import pandas as pd
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity

# --- 1. Base de datos (tu DataFrame consolidado) ---
df = pd.DataFrame(convocatorias)

# --- 2. Crear embeddings ---
model = SentenceTransformer("all-MiniLM-L6-v2")

# Usamos título + descripción como texto representativo
texts = df["titulo_convocatoria"] + " " + df["objetivo_descripcion"].fillna("")
embeddings = model.encode(texts.tolist())

# --- 3. Función de consulta ---
def consultar_llm(pregunta, top_k=3):
    pregunta_emb = model.encode([pregunta])
    sims = cosine_similarity(pregunta_emb, embeddings)[0]
    indices = sims.argsort()[-top_k:][::-1]

    resultados = df.iloc[indices]
    return resultados

# --- 4. Ejemplo de uso ---
pregunta = "desarrollo sostenible"
res = consultar_llm(pregunta)

print("Resultados relevantes:\n")
for _, row in res.iterrows():
    print("Institución:", row["institucion"])
    print("Título:", row["titulo_convocatoria"])
    print("Descripción:", row["objetivo_descripcion"])
    print("Fechas:", row["fecha_inicio"], "-", row["fecha_cierre"])
    print("Recursos:", row["recursos"])
    print("Link:", row["link"])
    print("="*80)
```

```python
import pandas as pd
import ipywidgets as widgets
from IPython.display import display
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity

# --- Base de datos ---
df = pd.DataFrame(convocatorias)

# --- Embeddings ---
model = SentenceTransformer("all-MiniLM-L6-v2")
texts = df["titulo_convocatoria"] + " " + df["objetivo_descripcion"].fillna("")
embeddings = model.encode(texts.tolist())

# --- Historial ---
historial = []

# --- Función de respuesta ---
def responder(pregunta, top_k=2):
    historial.append(("usuario", pregunta))

    pregunta_emb = model.encode([pregunta])
    sims = cosine_similarity(pregunta_emb, embeddings)[0]
    indices = sims.argsort()[-top_k:][::-1]
    resultados = df.iloc[indices]

    respuesta = "He encontrado estas convocatorias relacionadas:\n\n"
    for _, row in resultados.iterrows():
        respuesta += f"- {row['institucion']}: {row['titulo_convocatoria']}\n"
        if row['objetivo_descripcion']:
            respuesta += f"  Objetivo: {row['objetivo_descripcion']}\n"
        if row['fecha_inicio'] or row['fecha_cierre']:
            respuesta += f"  Fechas: {row['fecha_inicio']} - {row['fecha_cierre']}\n"
        if row['recursos']:
            respuesta += f"  Recursos: {row['recursos']}\n"
        respuesta += f"  Más info: {row['link']}\n\n"

    historial.append(("asistente", respuesta))
    return respuesta

# --- Widgets estilo Chat ---
entrada = widgets.Text(
    value='',
    placeholder='Escribe tu pregunta aquí...',
    description='Prompt:',
    layout=widgets.Layout(width='80%')
)

salida = widgets.Output()

# --- Botón para limpiar historial ---
boton_limpiar = widgets.Button(
    description="Limpiar historial",
    button_style="danger"
)

def mostrar_historial():
    salida.clear_output()
    with salida:
        for rol, contenido in historial:
            if rol == "usuario":
                display(widgets.HTML(f"<div style='background-color:#DCF8C6; padding:10px; border-radius:10px; margin:5px; text-align:right'><b>Tú:</b> {contenido}</div>"))
            else:
                display(widgets.HTML(f"<div style='background-color:#E6E6E6; padding:10px; border-radius:10px; margin:5px; text-align:left'><b>Asistente:</b> {contenido}</div>"))

def on_submit(change):
    if change.new.strip():
        responder(change.new)
        mostrar_historial()
        entrada.value = ""  # limpiar campo de texto

def limpiar_historial(b):
    historial.clear()
    mostrar_historial()

entrada.observe(on_submit, names="value")
boton_limpiar.on_click(limpiar_historial)

display(entrada, boton_limpiar, salida)

```
