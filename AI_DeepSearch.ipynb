# Ministerio de Ciencia Tecnología e Innovación

```python
def data_minciencias():
    base_url = "https://minciencias.gov.co"
    url_objetivo = "https://minciencias.gov.co/convocatorias/todas"  
    print(f"{'='*20} MINCIENCIAS: {'='*20}")
    try:
        resp = requests.get(url_objetivo, timeout=10)
        soup = BeautifulSoup(resp.text, "html.parser")
        datos_minciencias = []
        filas = soup.find_all("tr")
        for fila in filas:
            cols = fila.find_all("td")
            if len(cols) >= 2:
                link_tag = cols[1].find("a")
                if link_tag and link_tag.has_attr('href'):
                    url_final = urljoin(base_url, link_tag['href'])
                    titulo = link_tag.get_text(strip=True)
                    print(f"Institucion: MinCiencias")
                    print(f"Titulo: {titulo}")
                    print(f"Url: {url_final}")
                    print("-" * 50)                  
                    datos_minciencias.append({"institucion": "MinCiencias", "titulo_convocatoria": titulo, "url_externa": url_final})       
        return datos_minciencias
    except Exception as e:
        print(f"Error al corregir MinCiencias: {e}")
        return []
minciencias = data_minciencias()
```

# Fundación Natura Colombia

```python
def data_natura():
    url_objetivo = "https://natura.org.co/convocatorias/"
    base_url = "https://natura.org.co"  
    print(f"{'='*20} FUNDACIÓN NATURA: {'='*20}") 
    try:
        headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36'}
        resp = requests.get(url_objetivo, headers=headers, timeout=10)
        soup = BeautifulSoup(resp.text, "html.parser")
        datos_natura = []
        enlaces_vistos = set()
        for a in soup.find_all("a", href=True):
            href = a['href']
            if "convocatoria" in href.lower() and href not in enlaces_vistos:
                url_final = urljoin(base_url, href)
                slug = urlparse(url_final).path.strip('/')
                titulo_limpio = slug.replace('-', ' ').lower()
                titulo_limpio = titulo_limpio.replace('convocatoria', '').strip().capitalize()       
                print(f"Institucion: Fundación Natura")
                print(f"Titulo (Decodificado): {titulo_limpio}")
                print(f"Url: {url_final}")
                print("-" * 50)
                datos_natura.append({"institucion": "Fundación Natura", "titulo_convocatoria": titulo_limpio, "url_externa": url_final})
                enlaces_vistos.add(href)
        return datos_natura
    except Exception as e:
        print(f"Error al corregir Fundación Natura: {e}")
        return []
fundacion_natura = data_natura()
```

# Fundación para la Conservación y el Desarrollo

```python
def data_fcds():
    url_objetivo = "https://fcds.org.co/oportunidades-laborales/"
    base_url = "https://fcds.org.co"
    print(f"{'='*20} FCDS: {'='*20}")
    try:
        headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36'}
        resp = requests.get(url_objetivo, headers=headers, timeout=10)
        soup = BeautifulSoup(resp.text, "html.parser")     
        datos_fcds = []
        enlaces_vistos = set()
        enlaces = soup.find_all("a", href=True)
        for a in enlaces:
            href = a['href']
            texto_boton = a.get_text(strip=True)
            if (".pdf" in href.lower() and "uploads" in href.lower()) or ("más información" in texto_boton.lower()):        
                if href not in enlaces_vistos:
                    url_final = urljoin(base_url, href)
                    nombre_archivo = urlparse(url_final).path.split('/')[-1]
                    titulo_decodificado = nombre_archivo.replace('.pdf', '').replace('-', ' ').replace('_', ' ')
                    titulo_decodificado = titulo_decodificado.replace('tdr', 'TDR:').strip().capitalize()          
                    print(f"Institucion: FCDS")
                    print(f"Titulo: {titulo_decodificado}")
                    print(f"Url: {url_final}")
                    print("-" * 50)
                    datos_fcds.append({"institucion": "FCDS","titulo_convocatoria": titulo_decodificado,"url_externa": url_final})
                    enlaces_vistos.add(href)
        return datos_fcds
    except Exception as e:
        print(f"Error al corregir FCDS: {e}")
        return []
fcds = data_fcds()
```

# Patrimonio Natural

```python
def corregir_patrimonio_natural():
    url_objetivo = "https://patrimonionatural.org.co/convocatorias/"
    base_url = "https://patrimonionatural.org.co"
    
    print(f"{'='*20} PATRIMONIO NATURAL: {'='*20}")
    
    try:
        driver.get(url_objetivo)
        
        # Esperamos a que carguen los títulos principales (suelen ser h2 o h3)
        WebDriverWait(driver, 15).until(
            EC.presence_of_element_located((By.CLASS_NAME, "elementor-heading-title"))
        )
        
        soup = BeautifulSoup(driver.page_source, "html.parser")
        datos_patrimonio = []
        enlaces_vistos = set()

        # Buscamos los títulos que suelen tener la clase de Elementor
        titulos = soup.find_all(["h2", "h3"], class_="elementor-heading-title")

        for tag in titulos:
            link_tag = tag.find("a")
            if link_tag and link_tag.has_attr('href'):
                titulo = link_tag.get_text(strip=True)
                url_real = urljoin(base_url, link_tag['href'])
                
                # Filtramos para que el título sea relevante y la URL no sea una imagen
                if len(titulo) > 10 and not url_real.lower().endswith(('.jpg', '.png', '.jpeg')):
                    if url_real not in enlaces_vistos:
                        print(f"Institucion: Patrimonio Natural")
                        print(f"Titulo: {titulo}")
                        print(f"Url: {url_real}")
                        print("-" * 50)
                        
                        datos_patrimonio.append({
                            "institucion": "Patrimonio Natural",
                            "titulo_convocatoria": titulo,
                            "url_externa": url_real
                        })
                        enlaces_vistos.add(url_real)
        
        return datos_patrimonio

    except Exception as e:
        print(f"Error al corregir Patrimonio Natural: {e}")
        return []

# Ejecutamos la validación para la novena y última
data_9 = corregir_patrimonio_natural()
```

# Banco de Desarrollo de América Latina (CAF)

```python
def extraer_caf_final():
    # Definimos las rutas confirmadas
    urls_objetivo = [
        "https://www.caf.com/es/trabaja-con-nosotros/convocatorias/",
        "https://www.caf.com/es/trabaja-con-nosotros/licitaciones/"
    ]
    base_url = "https://www.caf.com"
    
    print(f"{'='*20} CAF: (Convocatorias + Licitaciones) {'='*20}")
    
    datos_caf = []
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36'}

    for url in urls_objetivo:
        try:
            resp = requests.get(url, headers=headers, timeout=15)
            soup = BeautifulSoup(resp.text, "html.parser")
            
            # Buscamos todos los H3 que contienen títulos según tus capturas
            # Seleccionamos tanto la clase de Convocatorias como la de Licitaciones
            encabezados_titulo = soup.find_all("h3", class_=["card__title", "card--gris-50__title"])
            
            for h3 in encabezados_titulo:
                link_tag = h3.find("a", href=True)
                
                if link_tag:
                    titulo = link_tag.get_text(strip=True)
                    href = link_tag['href']
                    url_final = urljoin(base_url, href)
                    
                    # Filtro de Seguridad y Relevancia (Habeas Data)
                    if len(titulo) > 10 and "mailto:" not in url_final.lower():
                        # Limpieza de duplicados de texto (ej. si el título trae códigos de licitación)
                        print(f"Institucion: CAF")
                        print(f"Titulo: {titulo}")
                        print(f"Url: {url_final}")
                        print("-" * 50)
                        
                        datos_caf.append({
                            "Institucion": "CAF",
                            "Titulo": titulo,
                            "Url": url_final
                        })
                        
        except Exception as e:
            print(f"Error procesando {url}: {e}")
            
    return datos_caf

# Ejecución para la base de datos de carlosmendez1997col
data_caf_total = extraer_caf_final()
```

# GIZ (Cooperación Alemana)

```python
def extraer_giz():
    # URL directa a las licitaciones internacionales (Tenders)
    url_objetivo = "https://www.giz.de/en/workingwithgiz/tenders.html"
    base_url = "https://www.giz.de"
    
    print(f"{'='*20} GIZ: {'='*20}")
    
    try:
        headers = {
            'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36'
        }
        resp = requests.get(url_objetivo, headers=headers, timeout=15)
        soup = BeautifulSoup(resp.text, "html.parser")
        
        datos_giz = []
        
        # En la GIZ, las licitaciones suelen estar dentro de listas o tablas 
        # Buscamos elementos <a> que contengan palabras clave de licitación
        enlaces = soup.find_all("a", href=True)

        for link in enlaces:
            href = link['href']
            # Filtramos enlaces que apunten a detalles de "tender" (licitación)
            if any(term in href.lower() for term in ["tender", "procurement", "contract"]):
                titulo = link.get_text(strip=True)
                url_final = urljoin(base_url, href)
                
                # --- ESTÁNDAR Y FILTRO ÉTICO ---
                # Evitamos ruido (enlaces de navegación genéricos)
                ruido = ["read more", "click here", "details", "contact", "download"]
                
                if len(titulo) > 20 and not any(r in titulo.lower() for r in ruido):
                    if "mailto:" not in url_final.lower():
                        print(f"Institucion: GIZ")
                        print(f"Titulo: {titulo}")
                        print(f"Url: {url_final}")
                        print("-" * 50)
                        
                        datos_giz.append({
                            "Institucion": "GIZ",
                            "Titulo": titulo,
                            "Url": url_final
                        })
        
        return datos_giz

    except Exception as e:
        print(f"Error al procesar GIZ: {e}")
        return []

# Ejecución
data_giz = extraer_giz()
```

```python
import os
import time
import getpass
import pandas as pd
from sqlalchemy import create_engine, text
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager

# --- 1. CONFIGURACIÓN DE CREDENCIALES SEGURAS ---
print("🔐 Configuración de conexión PostgreSQL Local")
DB_USER = input("Introduce el usuario de Postgres (ej. postgres): ")
DB_PASS = getpass.getpass("Introduce la contraseña de Postgres: ")
DB_NAME = input("Introduce el nombre de la base de datos (debe existir): ")
DB_HOST = "localhost"
DB_PORT = "5432"

# Crear el motor de conexión
try:
    engine = create_engine(f'postgresql://{DB_USER}:{DB_PASS}@{DB_HOST}:{DB_PORT}/{DB_NAME}')
    # Probar conexión inicial
    with engine.connect() as conn:
        print(f"✅ Conexión exitosa a la base de datos: {DB_NAME}")
except Exception as e:
    print(f"❌ Error de conexión: {e}")
    raise

# --- 2. PROCESO SIMULTÁNEO (NAVEGACIÓN + INYECCIÓN SQL) ---
def proceso_vivo_sql_seguro(df):
    tmp_dir = "temp_html_live"
    if not os.path.exists(tmp_dir): os.makedirs(tmp_dir)
    
    archivo_sql_respaldo = "datos_convocatorias_respaldo.sql"

    # Selenium VISIBLE
    chrome_options = Options()
    chrome_options.add_argument("--start-maximized")
    driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=chrome_options)

    # Asegurar que la tabla existe
    with engine.connect() as conn:
        conn.execute(text("CREATE TABLE IF NOT EXISTS licitaciones (institucion TEXT, titulo TEXT, url TEXT, informacion TEXT);"))
        conn.commit()

    print(f"\n🚀 Iniciando extracción masiva... Puedes monitorear la tabla 'licitaciones' en {DB_NAME}")
    
    

    try:
        with open(archivo_sql_respaldo, "w", encoding='utf-8') as f_sql:
            for idx, row in df.iterrows():
                try:
                    # A. Navegación Visible
                    driver.get(row['Url'])
                    time.sleep(5) 
                    html_raw = driver.page_source
                    
                    # B. Preparación de datos
                    # Escapamos comillas simples para la sentencia SQL
                    info_sql = html_raw.replace("'", "''")
                    inst = str(row['Institucion']).replace("'", "''")
                    tit = str(row['Titulo']).replace("'", "''")
                    url = str(row['Url']).replace("'", "''")

                    # C. Guardado temporal físico
                    path_html = os.path.join(tmp_dir, f"{idx}.html")
                    with open(path_html, "w", encoding='utf-8') as f:
                        f.write(html_raw)

                    # D. Inyección inmediata a Postgres
                    query = text(f"INSERT INTO licitaciones VALUES ('{inst}', '{tit}', '{url}', '{info_sql}');")
                    with engine.connect() as conn:
                        conn.execute(query)
                        conn.commit() # Commit inmediato para ver los datos en vivo
                    
                    # E. Escritura en archivo SQL de respaldo
                    f_sql.write(f"INSERT INTO licitaciones VALUES ('{inst}', '{tit}', '{url}', '{info_sql}');\n")

                    print(f"📥 [Registro {idx}] Inyectado en Postgres exitosamente.")

                except Exception as e:
                    print(f"⚠️ Error en registro {idx}: {e}")

    finally:
        driver.quit()

    # --- 3. LIMPIEZA FINAL DE HTML ---
    print("\n🗑 Borrando archivos temporales...")
    for file in os.listdir(tmp_dir):
        os.remove(os.path.join(tmp_dir, file))
    os.rmdir(tmp_dir)
    print(f"✨ Proceso terminado. Datos disponibles en Postgres y en {archivo_sql_respaldo}")

# --- EJECUCIÓN ---
# Nota: Lanza esto con tu DataFrame 'df_final_final' ya consolidado
proceso_vivo_sql_seguro(df_final_final)
```

```python
import pandas as pd
import ipywidgets as widgets
from IPython.display import display
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity

# --- Base de datos ---
df = pd.DataFrame(convocatorias)

# --- Embeddings ---
model = SentenceTransformer("all-MiniLM-L6-v2")
texts = df["titulo_convocatoria"] + " " + df["objetivo_descripcion"].fillna("")
embeddings = model.encode(texts.tolist())

# --- Historial ---
historial = []

# --- Función de respuesta ---
def responder(pregunta, top_k=2):
    historial.append(("usuario", pregunta))

    pregunta_emb = model.encode([pregunta])
    sims = cosine_similarity(pregunta_emb, embeddings)[0]
    indices = sims.argsort()[-top_k:][::-1]
    resultados = df.iloc[indices]

    respuesta = "He encontrado estas convocatorias relacionadas:\n\n"
    for _, row in resultados.iterrows():
        respuesta += f"- {row['institucion']}: {row['titulo_convocatoria']}\n"
        if row['objetivo_descripcion']:
            respuesta += f"  Objetivo: {row['objetivo_descripcion']}\n"
        if row['fecha_inicio'] or row['fecha_cierre']:
            respuesta += f"  Fechas: {row['fecha_inicio']} - {row['fecha_cierre']}\n"
        if row['recursos']:
            respuesta += f"  Recursos: {row['recursos']}\n"
        respuesta += f"  Más info: {row['link']}\n\n"

    historial.append(("asistente", respuesta))
    return respuesta

# --- Widgets estilo Chat ---
entrada = widgets.Text(
    value='',
    placeholder='Escribe tu pregunta aquí...',
    description='Prompt:',
    layout=widgets.Layout(width='80%')
)

salida = widgets.Output()

# --- Botón para limpiar historial ---
boton_limpiar = widgets.Button(
    description="Limpiar historial",
    button_style="danger"
)

def mostrar_historial():
    salida.clear_output()
    with salida:
        for rol, contenido in historial:
            if rol == "usuario":
                display(widgets.HTML(f"<div style='background-color:#DCF8C6; padding:10px; border-radius:10px; margin:5px; text-align:right'><b>Tú:</b> {contenido}</div>"))
            else:
                display(widgets.HTML(f"<div style='background-color:#E6E6E6; padding:10px; border-radius:10px; margin:5px; text-align:left'><b>Asistente:</b> {contenido}</div>"))

def on_submit(change):
    if change.new.strip():
        responder(change.new)
        mostrar_historial()
        entrada.value = ""  # limpiar campo de texto

def limpiar_historial(b):
    historial.clear()
    mostrar_historial()

entrada.observe(on_submit, names="value")
boton_limpiar.on_click(limpiar_historial)

display(entrada, boton_limpiar, salida)

```
